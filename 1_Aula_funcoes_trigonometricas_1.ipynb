Para construção de funções trigonométricas, dessa forma, entender o conceito de seno e cosseno. Usando as definições:


*   Função Seno: Claro! Vamos falar sobre a função seno de maneira simples.

A função seno é uma função matemática fundamental na trigonometria. Ela é usada para descrever a relação entre os ângulos e os comprimentos dos lados de um triângulo retângulo. Imagine um círculo com raio 1, chamado de círculo unitário. Se você pegar um ângulo e traçar uma linha do centro do círculo até a borda, formando um triângulo retângulo, o seno desse ângulo é a altura do triângulo quando você projeta a linha sobre o eixo horizontal.

Em termos mais simples, para um ângulo \(\theta\), o seno é a distância vertical do ponto onde a linha toca a borda do círculo até o eixo horizontal. Essa distância pode ser positiva ou negativa, dependendo de onde a linha toca o círculo.

A função seno varia entre -1 e 1. Quando o ângulo é 0 graus (ou 0 radianos), o seno é 0. À medida que o ângulo aumenta, o seno sobe até 1, depois desce até -1, e esse padrão se repete. Esse comportamento cria uma forma de onda suave e contínua, que é útil para modelar muitas situações, como o som e as ondas de luz.

*   Função Cosseno:
A função cosseno é outra função fundamental na trigonometria, muito parecida com a função seno. Em um círculo unitário, para um ângulo \(\theta\), o cosseno é a distância horizontal do ponto onde a linha que forma o ângulo toca a borda do círculo até o eixo vertical. Em termos mais simples, é a projeção da linha sobre o eixo horizontal.

A função cosseno, assim como o seno, varia entre -1 e 1. Quando o ângulo é 0 graus (ou 0 radianos), o cosseno é 1. À medida que o ângulo aumenta, o cosseno diminui até -1 e depois sobe de volta a 1, criando uma forma de onda suave e contínua. Esse padrão se repete, e o cosseno é útil para modelar fenômenos periódicos, como ondas e oscilações. Em resumo, a função cosseno ajuda a descrever como a distância horizontal de um ponto varia com o ângulo.



# Gráficos e suas construções usando

Os códigos abaixo são utilizados para montar o comportamento da função seno do tipo f(x)=a+sen(bx+c), em que, as constantes a,b e c mudam o comportamento da função seno e assim podemos determinar suas variações dependendo da aplicação que é utilizada. Nosso foco aqui é determinar os gráficos usando as bibliotecas Matplotlib e Numpy com códigos simples de construção gráfica.


```python
# prompt: Como coloco label no grafico no formado de multiplos de pi

import matplotlib.pyplot as plt
import numpy as np

# ... (your existing code for plotting) ...

# Customize x-axis labels
x_values = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
            ['$-2\pi$', '$-\pi$', '$0$', '$\pi$', '$2\pi$'])
plt.title('Funções trigonométricas')
plt.plot(x_values,np.sin(x_values), color = "black", linewidth = '2')
plt.plot(x_values,np.cos(x_values), color = "green", linewidth = '2')

plt.ylim(-1,1)
plt.grid()
plt.xlabel(r'$\theta$')
plt.ylabel(r'f($\theta)$')
plt.savefig('Funcoes_seno_cos_1_.png')


```
